```python
## we created class here because we want to store all function at one place
# 1.to determine weight and bias
# 2.forward pass= z = w*x+b and sigmoid
# 3.


class MySimpleNN():        ###

  def __init__(self,X):
    self.weights = torch.rand(X.shape[1],1,dtype = torch.float64,requires_grad = True)
    self.bias = torch.rand(1,dtype = torch.float64,requires_grad = True)

# forward pass
  def forward_pass(self,X):
    z = torch.matmul(X,self.weights) + self.bias
    y_pred = torch.sigmoid(z)
    return y_pred


    ## loss_function
# Agar output 0/1, ya Yes/No, ya class labels ho → Entropy-based loss
# Agar output real number ho (jaise weight, price, temperature) → MSE
# Breast Cancer dataset = 0 or 1
# → so “entropy (cross entropy / BCE)” automatically correct choice.
  def loss_function(self,y_pred,y):
     # Clamp predictions to avoid log(0)
    epsilon = 1e-7     ## ek chota sa number set kiya jo zero ke bahut paas hai
    y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)
    # torch.clamp(y_pred, epsilon, 1 - epsilon)
    # agar y_pred 0 tha → 0.0000001 bana diya
    # agar y_pred 1 tha → 0.9999999 bana diya
    # Taaki log(0) ya division-by-zero error na aaye


    # Calculate loss
# formula of loss    # loss = -(y * log(y_pred) + (1 - y) * log(1 - y_pred)).mean()

    loss = -(y_train_tensor * torch.log(y_pred) + (1 - y_train_tensor) * torch.log(1 - y_pred)).mean()
    return loss
```
