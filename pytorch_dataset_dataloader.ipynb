```python
import torch
from sklearn.datasets import make_classification        ## jab hum apni taraf se koi dataset bnaye just seekhne ke liye ya jb dataset available na ho it is dummy but looks like realistic
```

```python
# Step 1: Create a synthetic classification dataset using sklearn
X, y = make_classification(
    n_samples=10,       # Number of samples
    n_features=2,       # Number of features
    n_informative=2,    # Number of informative features
    n_redundant=0,      # Number of redundant features
    n_classes=2,        # Number of classes
    random_state=42     # For reproducibility
)
```

```python
# Convert the data to PyTorch tensors
X = torch.tensor(X, dtype=torch.float32)
y = torch.tensor(y, dtype=torch.long)            # y is not float i it is int but in pytorch we cannt use int64 we use torch.long = int64
```

```python
from torch.utils.data import Dataset, DataLoader
```

```python
class CustomDataset(Dataset):
  def __init__(self,features,labels):      ## isko input mein features and labels milenge
    self.features = features
    self.labels = labels

  def __len__(self):
    return self.features.shape[0]        ## input mein kuch ni milega but ye return karega length of features

  def __getitem__(self,index):    ## isme index denge to ye wahi row nikal ke dega
    return self.features[index],self.labels[index]
```

```python
dataset = CustomDataset(X,y)    ## object bna liya class se
```

```python
len(dataset)
```

```python
dataset[2]
```

```python
dataloader = DataLoader(dataset, batch_size=2, shuffle=False)   ## shuffle false kiya means ki sequentially har 2-2 ko milake ek batchsize banega
```

```python
for batch_features,batch_labels in dataloader:
  print(batch_features)
  print(batch_labels)
  print("-"*50)   ## ye ek boundary hai jo har batch ke baad create hui hai
```
