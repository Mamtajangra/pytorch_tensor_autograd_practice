```python
class Model(nn.Module):
  def __init__(self,num_features):

    super().__init__()
    self.network = nn.Sequential(
        nn.Linear(num_features,3),
    nn.ReLU(),
    nn.Linear(3,1),
    nn.Sigmoid()

    )

  def forward(self,features):
      out = self.network(features)

      return out
```

# we have 5 features here means 5 columns and after that we have 3 output in hidden layer and at outermost there are 1 output means 5 features = 15 weights and 3 bias again 3 weights = 1 bias

```python
# create fake dataset
features = torch.rand(10,5)

# create model
model = Model(features.shape[1])

output = model(features)
print(output)
```
